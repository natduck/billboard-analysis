```python
def get_genre(artist, track):
    """ gets the genre of the artist and their track
    
    args:
        artist(str): artist
        track(str): track name
        
    returns: 
        genres(list): list of genres or 'NO GENRE' if response is None
    """
    
    # Set up the Last.fm API endpoint and parameters
    api_key = '99ada5f533810124997231207a4df132' # Replace with your Last.fm API key
    base_url = 'http://ws.audioscrobbler.com/2.0/'
    method = 'track.getInfo'
    
    # create parameters
    params = {
    'method': method,
    'api_key': api_key,
    'artist': artist,
    'track': track,
    'format': 'json'
    }
    
    # Send the API request and get the response
    response = requests.get(base_url, params=params)
    data = json.loads(response.text)

    # Extract the genres from the response
    if data.get('error'):
        genres = 'NO GENRE'
    else:
        tags = data['track']['toptags']['tag']
        genres = [tag['name'] for tag in tags]
    
    return genres
```

```python
# appending lyrics to the dataset

from profanityfilter import ProfanityFilter
import nltk
from nltk.corpus import stopwords




def clean_lyrics(string):
    """ cleans the lyrics """
    string = string.replace("*", "")
    string = string.replace(",", "")
    string = string.replace("!", "")
    string = string.replace("?", "")
    string = string.replace("'", "")
    string = string.replace("(", "")
    string = string.replace(")", "")
    string = string.replace("-", "")
    
    return string

def lower_song(lyrics):
    """ returns the lyrics in lowercase 
    
    args:
        lyrics(str): the lyrics
        
    returns:
        lyrics(str): lowercased lyrics """
    
    lyrics = lyrics.lower()
    
    return lyrics


def remove_headers(lyrics):
    """ removes headings in songs
    
    args:
        lyrics(str): the lyrics
        
    returns:
        lyrics(str): lyrics with no heads """
    
    # Remove lines containing 'chorus', 'verse', or 'solo'
    lyrics = '\n'.join([line for line in lyrics.split('\n') if 'chorus' not in line.lower()\
                    and 'verse' not in line.lower()\
                    and 'solo' not in line.lower()\
                    and 'intro' not in line.lower()\
                    and 'skit' not in line.lower()\
                    and 'translation' not in line.lower()])
    
    return lyrics

def clean_string(string):
    """ cleans the string and removes punctuation and makes into one large string with no whitespace
    
    args:
        string(str): string
    
    returns:
        string(str): cleaned string """
    # Convert the string to lowercase
    string = string.lower()

    # Remove all newlines and extra whitespaces
    string = ' '.join(string.split())

    # Remove all punctuations
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for char in string:
        if char in punctuations:
            string = string.replace(char, '')

    return string


def apply_lyrics(artist, song):
    """ use of Genius API to add lyrics that have stopwords removed 
    
    args:
        artist(str): artist
        song(str): song name 
        
    returns:
        lyric_list(list): cleaned, lowercase lyrics """
    
    # initalize API
    client_id = '96dcd39191cf46ee81bb5d1092e6716b'
    client_secret = '22a0463e07e2485c8fc0a692d3412183'

    client_credentials_manager = SpotifyClientCredentials(client_id, client_secret)
    sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
    


    # Access Genius API
    genius_api_key = 'iYtxgNLr8PhZsYubPUAEARAR_tWMscYA41pAv7YGZryANx9-eFThYcDS2pBfvOPoJVD_VsKbsOxlWwq9KUtyHw'
    # Access Genius API with increased timeout
    genius = lyricsgenius.Genius(genius_api_key, timeout=130)


    # search for the lyrics using the Genius API
    song = genius.search_song(song, artist)
    
    if song is None:
        return None
    lyrics = song.lyrics
    
    if lyrics is None:
        return None
    
    else:

        # censor lyrics
        pf = ProfanityFilter()
        lyrics = pf.censor(lyrics)

        # clean lyrics
        lyrics = clean_lyrics(lyrics)

        # make lowercase
        lyrics = lower_song(lyrics)

        # remove headers
        lyrics = remove_headers(lyrics)

        # clean
        lyrics = clean_string(lyrics)

        # initialize stopwords
        stops = set(stopwords.words('english'))

        # make lyrics into a list that doesnt have stopwords
        lyric_list = [word for word in lyrics.split() if word not in (stops)]
    
        return lyric_list
```

```python
def remove_digits_and_embed(lst):
    """
    Remove digits and the word 'embed' from the list elements and embed only the ones without digits or 'embed'
    to the last element of the list.

    args:
        lst (list): A list of strings.

    returns:
        list: A list of strings where the ones containing digits or 'embed' are removed and only the remaining ones
        are embedded to the last element of the list.

    """
    # Create a list of strings that do not contain digits or 'embed'
    non_digit_items = [item for item in lst if not any(char.isdigit() for char in item) and item != 'embed']

    # Concatenate the non-digit strings and assign them to a new string variable
    embedded_str = ''.join(non_digit_items)

    # Return the list of non-digit items and the concatenated string as the last element of the list
    return non_digit_items
```

```python
import nltk  # import the Natural Language Toolkit library
nltk.download('vader_lexicon')  # download the pre-trained VADER lexicon for sentiment analysis
from nltk.sentiment.vader import SentimentIntensityAnalyzer  # import the VADER sentiment analysis tool

def sentiment_score(text_list):
    """
    Compute the sentiment score of a list of strings using the VADER sentiment analysis tool.

    args:
        text_list (list): A list of strings to be analyzed for sentiment.

    returns:
        total_score(float): The sum of the compound sentiment scores of all the strings in the list.
    """
    sia = SentimentIntensityAnalyzer()  # create an instance of the VADER sentiment analyzer
    total_score = 0.0  # initialize a variable to store the total sentiment score
    for text in text_list:  # iterate over each string in the input list
        scores = sia.polarity_scores(text)  # get the sentiment scores for the current string using VADER
        total_score += scores['compound']  # add the compound score to the total score
    return total_score  # return the final total sentiment score
```

```python
df['Sentiment'] = df['Lyrics'].apply(sentiment_score)
```
